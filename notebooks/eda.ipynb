## EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/creditcard.csv")
```

## class imbalance 

```python
counts = df["Class"].value_counts().sort_index()

colors = ['blue', 'red']  # 0 blue, 1 red

counts.plot(kind="bar", color=colors)

plt.title("Class Distribution")
plt.xlabel("Class (0 = Normal, 1 = Fraud)")
plt.ylabel("Count")
plt.grid(axis='y', alpha=0.3)
plt.savefig("../reports/figures/Class_imbalance.png")
plt.show()
```

### Observation
- The dataset is highly Imbalnced, with very few fraud Cases.
- which may cause our model to Predice on luck
- Percent of each class :
    - 98% - 0 
    - 17% - 1

## Amount Distribution

```python
plt.figure(figsize=(8,4))

sns.histplot(
    df['Amount'],
    bins=50,
    kde=True
)

plt.title("Transaction Amount Distribution", fontsize=16)
plt.xlabel("Amount", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

plt.grid(alpha=0.3)
plt.savefig("../reports/figures/amount_distribution.png")
plt.show()
```

### Observation
- Explains that Amount is Highly Skewed with Most transiction Being of Small VAL

# TIME

```python
sns.histplot(data=df, x='Time', hue='Class', bins=50)
plt.title("Fraud vs Normal over Time")
plt.grid()
plt.show()
```

### Observation
- Fraud happenend Over Time 
- with fraud being only 0.2% it is basically Hided
- Non Fraud (0) is Dominating the Graph

# Correlation Heatmap 

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df.corr(),
    cmap="coolwarm",
    center=0,
    cbar=True,
    square=False
)

plt.title("Correlation Heatmap", fontsize=16)
plt.xticks(rotation=90)
plt.yticks(rotation=0)

plt.tight_layout()
plt.savefig("../reports/figures/correlation_heatmap.png")
plt.show()
```

### Observation
- Heatmap showing that most features having low correlation with each other
- high independent
- Feature relations are minimal
- This is Expected PCA() transformed
